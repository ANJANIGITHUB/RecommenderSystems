### Popularity Based Recommendation System

```python
#Recommend Popular Books in the platform
df_books_ratings=df_books.merge(df_ratings,on='ISBN')
```

```python
df_books_ratings.head()
```

```python
df_books_ratings.shape
```

### Get the number of books count

```python
#Get the number of books count
ratings_count=df_books_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
ratings_count.rename(columns={'Book-Rating':'Rating-Count'},inplace=True)
ratings_count.sort_values(by='Rating-Count',ascending=False)
```

### Get the avarage of books ratings

```python
#Get the avarage of books ratings
avg_count=df_books_ratings.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_count.rename(columns={'Book-Rating':'Rating-Average'},inplace=True)
avg_count.sort_values(by='Rating-Average',ascending=False)
```

```python
df_rating_avg=ratings_count.merge(avg_count,on='Book-Title')
```

```python
df_rating_avg
```

### Users with 250 & More Ratings

```python
df_rating_avg_grt_250=df_rating_avg.loc[df_rating_avg['Rating-Count']>250]
```

```python
popular_books_df=df_rating_avg_grt_250.sort_values(by='Rating-Average',ascending=False).head(50)
```

```python
popular_books_df.head()
```

### Colloborative Filtering Recommender System

```python
df_books_ratings.head()
```

### Finding the users who has done more then 100 ratings on the books

```python
users_with100_ratings=df_books_ratings.groupby('User-ID').count()['Book-Rating']>100
```

```python
usersids_with100_ratings=users_with100_ratings[users_with100_ratings].index
```

```python
usersdata_with100_ratings=df_books_ratings[df_books_ratings['User-ID'].isin(usersids_with100_ratings)]
```

```python
usersdata_with100_ratings.head()
```

### Finding the books which has got atleast 50 ratings

```python
books_users_data=usersdata_with100_ratings.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
books_names=books_users_data[books_users_data].index
```

```python
final_user_ratings=usersdata_with100_ratings[usersdata_with100_ratings['Book-Title'].isin(books_names)]
```

```python
final_user_ratings.drop_duplicates(inplace=True)
```

```python
final_user_ratings.head()
```

### Create User-Item Matrix using Pivot Table

```python
User_Item_IM=final_user_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
User_Item_IM.fillna(0,inplace=True)
```

```python
User_Item_IM[254]
```

### Finding the distance of each book with other books

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cos_sim=cosine_similarity(User_Item_IM)
```

### Find Five Similar Books

```python
def recommend(book_name):
    #get the index number of book
    index=np.where(User_Item_IM.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(cos_sim[index])),key=lambda x:x[1],reverse=True)[1:6]
    item=[]
    print("Similar Books to Book '{}' are below:".format(book_name))
    for i in similar_items:
        #item=User_Item_IM.index[i[0]]
        item.append(User_Item_IM.index[i[0]])
    return item

    
```

```python
recommend('Message in a Bottle')
```
